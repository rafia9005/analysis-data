```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
```

```python
df = pd.read_csv("data_real.csv")
df.head()
```

```python
label = []
for index, row in df.iterrows():
    if row["Rating"] == 1 or row["Rating"] == 2:
        label.append(0)
    else:
        label.append(1)

df["label"] = label
df.head()
```

```python
# change font  to lower
def change_case_folding(Review):
  Review = Review.lower()
  return Review
df["Ulasan"] = df["Ulasan"].apply(change_case_folding)

df.head()
```

```python
# normalisasi data
norm = {" dgn " : " dengan ", " gue " : " saya ", " bgmn " : " bagaimana ", " tdk " : " tidak ", " blum " :  "blum", " mamtaaaaaaaappp" : "bagus"}
def normalisasi_data(str_text):
  for i in norm:
    str_text = str_text.replace(i, norm[i])
  return str_text

df["Ulasan"] = df["Ulasan"].apply(lambda x: normalisasi_data(x))
df.head()
```
